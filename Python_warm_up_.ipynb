```python
import pandas as pd
```

```python
# Load the data into a dataframe (Excel spreadsheet alike object)

df = pd.read_csv('workshop_data.csv')
```

# Classification

- We want to predict an outcome (male,female / yes,no / a,b,c)
- target: what we want to predict?
- features / inputs: what we want to **USE** to predict?

```python
df
```

```python
# We configure the parameters / settings

target = "gender" # string
inputs = ["weight", "height"] # list
```

$ y = mx + c $

y is our target
X is our inputs

```python
# We prepare our TRAINING data
# Training data is used to train our ML model

X = df[inputs]
y = df[target]
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(X,y)
```

```python
model.predict([[90, 170], [50, 155]])
```

```python
import graphviz
columns = list(X.columns)

dot_data = tree.export_graphviz(model, out_file=None, feature_names=columns, class_names=['Male', 'Female'], filled = True, rounded = True)

graph = graphviz.Source(dot_data)
graph
```
